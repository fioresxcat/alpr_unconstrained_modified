#**2. Bắt đầu code**

```python
# Load model LP detection
wpod_net_path = "my_modified_model_final.json"
wpod_net = load_model(wpod_net_path)
```

```python
# Đường dẫn ảnh, các bạn đổi tên file tại đây để thử nhé
img_path = "test/vuong_92.jpg"

# Đọc file ảnh đầu vào
Ivehicle = cv2.imread(img_path)
cv2.imshow('', Ivehicle)
cv2.waitKey(0)

# Kích thước lớn nhất và nhỏ nhất của 1 chiều ảnh
Dmax = 608
Dmin = 288

# Lấy tỷ lệ giữa W và H của ảnh và tìm ra chiều nhỏ nhất
ratio = float(max(Ivehicle.shape[:2])) / min(Ivehicle.shape[:2])
side = int(ratio * Dmin)
bound_dim = min(side, Dmax) # cái này để resize lại ảnh trước khi đưa vào wpod-net đây

final_labels , LpImg, lp_type = detect_lp(wpod_net, im2single(Ivehicle), bound_dim, lp_threshold=0.5)
print(lp_type)

if (len(LpImg)):
    # Xử lý đọc biển đầu tiên, các bạn có thẻ sửa code để detect all biển số
    plt.imshow(LpImg[0])
    cv2.waitKey()
```
